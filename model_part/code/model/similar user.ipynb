```python
#merge all the dataset having skills from user profile.
dflang = df1[['Respondent','LanguageWorkedWith']]
dfdata = df1[['Respondent','DatabaseWorkedWith']]
dfplat = df1[['Respondent','PlatformWorkedWith']]
dfframe = df1[['Respondent','FrameworkWorkedWith']]
dfdev = df1[['Respondent','DevType']]
dfmerge=pd.merge(dflang,dfdata,on="Respondent")
dfmerge=pd.merge(dfmerge,dfplat,on="Respondent")
dfmerge=pd.merge(dfmerge,dfframe,on="Respondent")
dfmerge=pd.merge(dfmerge,dfdev,on="Respondent")
dfuser = dfmerge
```

```python
import numpy as np
from numpy import array
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

#remove duplicate columns during merge
dfuser = dfuser.loc[:,~dfuser.columns.duplicated()]
dfuser=dfuser.dropna().reset_index(drop=True)

#remove rows with less than 5 skills
dfuser=dfuser.dropna(thresh=5)
dfuser.shape


ps = PorterStemmer()
all_stopwords = stopwords.words('english')
skill_index = {1:'LanguageWorkedWith',2:'DatabaseWorkedWith',3:'PlatformWorkedWith',4:'FrameworkWorkedWith',5:'DevType'}

# preprocessing
for index,row in dfuser.iterrows():
    #skills = row[1].values[1:]
    skills  = row
    for j in range(1,len(skills)):
        temp = re.sub('[^a-zA-Z]', ' ', skills[j])
        temp = temp.lower()
        temp = re.split('[,/; ]',temp)
        temp = [ps.stem(word.strip()) for word in temp if not word in set(all_stopwords)]
        temp = ' '.join(temp)
        dfuser[skill_index[j]][index] = temp
```

```python
#convert job type into vectors 
tf = TfidfVectorizer()
f_matrix_jt = tf.fit_transform(dfuser_5000['job type'])
```

```python
f_matrix_jt
```

```python
scores = []
for row in dfuser_5000.iterrows():
    index,data = row
    #current respondent
    resp = data[0]
    jt_resp = dfuser_5000['job type'][index]
    l1 = len(jt_resp)
    
    #max_index[resp] is the most similar user to current resp
    simi_resp =  max_index[resp]
    print(resp,simi_resp)
    if simi_resp == 0: # not  recommend similar respondent to this user
        #scores.append(0)
        continue
    index_sim_resp = dfuser_5000[dfuser_5000['respondent'] == simi_resp].index.tolist()[0]
    
    #jt_sim_resp = dfuser_5000[dfuser_5000['respondent'] == simi_resp]['job type']
    jt_sim_resp = dfuser_5000['job type'][index_sim_resp]
    l2 = len(jt_sim_resp)
       
    if l1 < l2:
        if jt_resp in jt_sim_resp:
            scores.append(1)
        else:
            score = cosine_similarity(f_matrix_jt[index,:],f_matrix_jt[index_sim_resp,:])[0][0]
            if score >= 0.5:
                scores.append(1)
            else:
                scores.append(score)
    else:
        if jt_sim_resp in jt_resp:
            scores.append(1)
        else:
            score = cosine_similarity(f_matrix_jt[index,:],f_matrix_jt[index_sim_resp,:])[0][0]
            if score >= 0.5:
                scores.append(1)
            else:
                scores.append(score)
```
