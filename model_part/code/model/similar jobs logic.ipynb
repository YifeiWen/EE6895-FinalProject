```python
import re
import sys
import getopt
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem.porter import PorterStemmer
import joblib
import pdfplumber
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path = '../../data/job_data.csv'
jobs = pd.read_csv(path)
jobs.head(5)
```

```python
#drop some rows without skills information
jobs = jobs.dropna(subset=['skills'])
jobs = jobs.drop(jobs[jobs['skills']=='SEE BELOW'].index)
jobs = jobs.reset_index()

for i in range(len(jobs)):
    jobs_temp = jobs['skills'][i]
    jobs['skills'][i] = jobs_temp.lower()
    
jobs = jobs[~jobs['skills'].str.contains('description')]
jobs = jobs[~jobs['skills'].str.contains('the')]
jobs = jobs[~jobs['skills'].str.contains('you')]
jobs = jobs.reset_index()
jobs.head(5)
```

```python
#get skills by countvector
ps = PorterStemmer()
jobs = joblib.load('../model_save/cleaned_jobs_data.pkl')
all_stopwords = stopwords.words('english')
for index,row in jobs.iterrows():
    temp = re.sub('[^a-zA-Z]', ' ', row['skills'])
    temp = temp.lower()
    temp = re.split('[,/;]',temp)
    temp = [ps.stem(word.strip()) for word in temp if not word in set(all_stopwords)]
    temp = ' '.join(temp)
    jobs['skills'][index] = temp
```
