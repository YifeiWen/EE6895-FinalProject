```python
import re
import sys
import getopt
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem.porter import PorterStemmer
```

```python
import pdfplumber
#path = 'C:/Users/wsy/Desktop/Resume/SDE/'
path = 'C:/Users/wsy/Desktop/'
with pdfplumber.open(path + 'fcccccc.pdf') as rpdf:
    first_page = rpdf.pages[0]
    cv_pdf = first_page.extract_text() 
```

```python
cv_pdf
```

```python
all_stopwords = stopwords.words('english')
```

```python
for i in range(len(cv_pdf)):
    if not cv_pdf[i].isalpha():
        cv_pdf = cv_pdf.replace(cv_pdf[i],' ')
cv_pdf = cv_pdf.lower().split()

ps = PorterStemmer()
temp = []
for j in range(len(cv_pdf)):
    if not cv_pdf[j] in all_stopwords:
        temp.append(ps.stem(cv_pdf[j]))
temp = ' '.join(temp)
cv_pdf = temp
print(cv_pdf)
```

```python
path = 'C:/Users/wsy/Desktop/ELEN6895/project/project/Data/processed_data.csv'
jobs = pd.read_csv(path)
print(jobs.head(5))
```

```python
tf = TfidfVectorizer()
tf_jobs = tf.fit_transform(jobs["Description"])
tf_cv_pdf = tf.transform([cv_pdf])
```

```python
similarity = cosine_similarity(tf_jobs, tf_cv_pdf)
print(similarity)
similarity[1][0]
```

```python
dataset = pd.read_csv("../Data/IT_Data.csv")
```

```python
#calculte the score of each category
#normalize score
category = dataset['Query'].unique()
similarity_score = {x:0 for x in category} 
for i in range(len(similarity)):
    similarity_score[jobs["Query"][i]] += similarity[i][0]


summation = sum(similarity)
for key,value in similarity_score.items():
    similarity_score[key] /= summation[0]

similarity_score
```

```python
similarity_index = np.zeros((len(similarity),2))
```

```python
#weigthed similarity score
for i in range(len(similarity)):
    similarity_index[i][0] = similarity[i][0] * similarity_score[jobs["Query"][i]]
    similarity_index[i][1] = i 
```

```python
sorted_similarity_index = sorted(similarity_index,key = (lambda x : x[0]),reverse = True)
```

```python
sorted_similarity_index
```

```python
dataset = pd.read_csv("../Data/IT_Data.csv")
dataset = dataset.drop(['ID','Job Title'],axis = 1)
print(dataset.head(5))
```

```python
k = 3
query = []
desp = []
for i in range(k):
    #print([jobs['Query'][sorted_similarity_index[i][1]],jobs['Description'][sorted_similarity_index[i][0]]])
    query.append(jobs['Query'][sorted_similarity_index[i][1]])
    desp.append(dataset['Description'][sorted_similarity_index[i][1]])
    #print(jobs['Query'][sorted_similarity_index[i][1]],dataset['Description'][sorted_similarity_index[i][1]])

res_df = pd.DataFrame(np.array([query,desp]).T,columns=['query','description'])
res_df.head(5)
```
